## rcParams

rcParams are settings for matplotlib. They're how stylesheets work (the thing when you use `plt.style.use`.

You can find examples of the different styles [here](https://matplotlib.org/2.1.1/gallery/style_sheets/style_sheets_reference.html) and [here](https://tonysyu.github.io/raw_content/matplotlib-style-gallery/gallery.html).

> **A big long list is:** bmh, classic, dark_background, fivethirtyeight, ggplot, grayscale, seaborn-bright, seaborn-colorblind, seaborn-dark, seaborn-dark-palette, seaborn-darkgrid, seaborn-deep, seaborn-muted, seaborn-notebook, seaborn-paper, seaborn-pastel, seaborn-poster, seaborn-talk, seaborn-ticks, seaborn-white, seaborn-whitegrid 

If you think one of them is cool, you can find its rcparams styles by checking out its mplstyle file in [in this directory](https://github.com/matplotlib/matplotlib/tree/master/lib/matplotlib/mpl-data/stylelib).

# Grouped bar graph (2 colors)

```python
df = pd.read_csv("gender.csv")
df.head(3)
```

```python
ax = df.plot(x='state', kind='barh', figsize=(10,4))

# CUSTOMIZATION CODE GOES UNDER HERE
ax.grid(False, axis='y')
ax.yaxis.label.set_visible(False)
ax.set_title("Cool Economist Graph 1", pad=20, loc='left', weight='bold', size='16')

# Get the figure even though you didn't do the subplots thing
fig = plt.gcf()
# Create a rectangle anchored at 40% of the way across
# and 50% of the way down. height/width grow UP and RIGHT.
rect = plt.Rectangle((0,1), height=0.13, width=0.015, transform=fig.transFigure, figure=fig, color='red')
# Add it to the figure
fig.patches.append(rect)

fig = plt.gcf()
rect = plt.Rectangle((0,1.13), height=0.00001, width=1, transform=fig.transFigure, figure=fig, color='red')
fig.patches.append(rect)

plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)

# then we save it
plt.savefig("graph-bar-grouped-economist.pdf")
plt.show()
```

# Bar graph (Single color)

```python
df = pd.read_csv("countries.csv")
result = df.groupby("continent").life_expectancy.median().sort_values()
result
```

```python
color = plt.rcParams['axes.prop_cycle'].by_key()['color'][0]
ax = result.plot(kind='barh', figsize=(10, 5), color=color)

# CUSTOMIZATION CODE GOES UNDER HERE
ax.spines['top'].set_color('red')
ax.spines['left'].set_color('red')
ax.grid(False, axis='y')
ax.yaxis.label.set_visible(False)
ax.set_title("Cool Economist Graph 2", pad=20, loc='left', weight='bold', size='16')

# Get the figure even though you didn't do the subplots thing
fig = plt.gcf()
# Create a rectangle anchored at 40% of the way across
# and 50% of the way down. height/width grow UP and RIGHT.
rect = plt.Rectangle((0,1), height=0.13, width=0.015, transform=fig.transFigure, figure=fig, color='red')
# Add it to the figure
fig.patches.append(rect)

fig = plt.gcf()
rect = plt.Rectangle((0,1.13), height=0.00001, width=1, transform=fig.transFigure, figure=fig, color='red')
fig.patches.append(rect)
# then we save it
plt.savefig("graph-bar-economist.pdf")
plt.show()
```

# Line graph

```python
df = pd.read_csv("visas-by-continent-long.csv")
df = df.groupby(["year", "continent"]).visas.sum().unstack().reset_index()
df.drop(columns='Unknown', inplace=True)
df['year'] = pd.to_datetime(df.year, format="%Y")
df.set_index('year', inplace=True)
percent_df = df.div(df.sum(axis=1), axis='rows') * 100

percent_df.head(2)
```

```python
ax = percent_df.plot(kind='line', figsize=(10, 5))

# CUSTOMIZATION CODE GOES UNDER HERE
ax.spines['top'].set_color('red')
ax.grid(False, axis='x')
ax.set_title("Cool Economist Graph 3", pad=65, loc='left', weight='bold', size='16')
ax.set_xlabel("Year")

# Get the figure even though you didn't do the subplots thing
fig = plt.gcf()
# Create a rectangle anchored at 40% of the way across
# and 50% of the way down. height/width grow UP and RIGHT.
rect = plt.Rectangle((0,1), height=0.13, width=0.015, transform=fig.transFigure, figure=fig, color='red')
# Add it to the figure
fig.patches.append(rect)

fig = plt.gcf()
rect = plt.Rectangle((0,1.13), height=0.00001, width=1, transform=fig.transFigure, figure=fig, color='red')
fig.patches.append(rect)

plt.legend(bbox_to_anchor=(0., 1.02, 1., .102), loc=3,
           ncol=2, mode="expand", borderaxespad=0.)

# then we save it
plt.savefig("graph-lines-economist.pdf")
plt.show()
```

# Scatter plot

```python
df = pd.read_csv("countries.csv")
df.head(2)
```

```python
fig, ax = plt.subplots(figsize=(8, 6))

# Don't touch this part of the code!
colors = list(reversed(plt.rcParams['axes.prop_cycle'].by_key()['color']))

groups = df.groupby('continent')
for name, group in groups:
    group.plot(kind='scatter', x='gdp_per_capita', y='life_expectancy', color=colors.pop(), label=name, s= 5 + (group.population / 200000), ax=ax, alpha=0.75)
ax.legend_.remove()

# CUSTOMIZATION CODE GOES UNDER HERE
ax.spines['top'].set_color('red')
ax.grid(False, axis='x')
ax.set_title("Cool Economist Graph 4", pad=20, loc='left',weight='bold', size='16')
ax.set_xlabel("GDP Per Capita")
ax.set_ylabel("Life Expectancy")

# Get the figure even though you didn't do the subplots thing
fig = plt.gcf()
# Create a rectangle anchored at 40% of the way across
# and 50% of the way down. height/width grow UP and RIGHT.
rect = plt.Rectangle((0,1), height=0.13, width=0.015, transform=fig.transFigure, figure=fig, color='red')
# Add it to the figure
fig.patches.append(rect)

fig = plt.gcf()
rect = plt.Rectangle((0,1.13), height=0.00001, width=1, transform=fig.transFigure, figure=fig, color='red')
fig.patches.append(rect)

# then we save it
plt.savefig("graph-scatter-economist.pdf")
plt.show()
```
